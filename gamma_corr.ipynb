```python
import cv2
img = cv2.imread("rosbag_imgs/yellow.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.imshow(img)
```

```python
import numpy as np
```

```python
gamma = 0.3
invGamma = 1.0 / gamma
table = np.array([((i / 255.0) ** invGamma) * 255 for i in np.arange(0, 256)]).astype("uint8")

img2 = cv2.LUT(img, table)
```

```python
plt.imshow(img)
plt.show()
plt.imshow(img2)
```

```python
img_red = cv2.imread("rosbag_imgs/red.png")
img_red = cv2.cvtColor(img_red, cv2.COLOR_BGR2RGB)
plt.imshow(img_red)
```

```python
plt.imshow(img_red2)
```
